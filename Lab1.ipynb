```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftshift

# Parameters
Fs = 8000  # Sampling Frequency
T = 1/Fs  # Sampling Interval
t = np.arange(0, 0.05, T)  # Time Vector
x = 3 + 2 * np.cos(120 * np.pi * t)  # Sinusoidal Signal
# Plot Signal in Time Domain
plt.figure(figsize=(17, 4))
plt.subplot(2, 1, 1)  # Prepare subplot
plt.plot(t, x, linewidth=2)
plt.title('Signal in Time Domain')
plt.xlabel('Time [s]')
plt.xticks(np.arange(0, 0.05, 0.005))
plt.ylabel('Amplitude')

# Fast Fourier Transform (Magnitude)
X = np.abs(fftshift(fft(x))) / len(x)

# Frequency Variable
N = len(x)
f = (np.arange(-N/2, N/2) * Fs) / N

# Plot Signal in Frequency Domain
# cambiar tama√±o 
plt.figure(figsize=(13, 4))
plt.subplot(2, 1, 2)  # Prepare subplot
plt.plot(f, X, linewidth=2)
plt.title('Signal in Frequency Domain')
plt.xlabel('Frequency [Hz]')
plt.xticks(np.arange(-4000, 4000, 100))
# rotar los valores del eje x
plt.xticks(rotation=90)
plt.ylabel('Magnitude')

plt.tight_layout()  # Adjust the layout of the plots
plt.show()  # Display the plots
```

La relaci√≥n entre las **componentes espectrales** de una se√±al y su representaci√≥n en el **dominio del tiempo** es fundamental para entender c√≥mo se compone una se√±al en t√©rminos de sus frecuencias. Las componentes espectrales de una se√±al son una representaci√≥n de la misma en el dominio de la frecuencia, lo que nos permite ver las frecuencias que componen la se√±al original y la amplitud de estas frecuencias.

En el ejemplo que analizamos, la se√±al en el dominio del tiempo es una suma de un t√©rmino constante y un coseno, espec√≠ficamente:

$$x(t) = 3 + 2 \cos(120\pi t)$$


Esta ecuaci√≥n muestra que la se√±al es una combinaci√≥n de una componente de **corriente continua (DC)** debido al t√©rmino "3" y una componente **sinusoidal** con una amplitud de "2" y una frecuencia de oscilaci√≥n de $$120\pi$$ radianes por segundo. Al convertir esta frecuencia a hertz, consideramos que $$f = \frac{\omega}{2\pi}$$, lo que nos da una frecuencia de **60 Hz** para la componente sinusoidal.

Por lo tanto, en el **espectro de frecuencias** de esta se√±al, esperar√≠amos ver:

1. **Componente de DC**: Una delta (pico) en **0 Hz** correspondiente al t√©rmino constante de la se√±al. Esto se debe a que una se√±al constante en el tiempo se interpreta como una se√±al de frecuencia cero en el dominio de la frecuencia.

2. **Componentes sinusoidales**: Picos en **¬±60 Hz** correspondientes a la componente sinusoidal de la se√±al. Estos picos representan la frecuencia fundamental de la se√±al sinusoidal y su inversa, dado que la transformada de Fourier considera componentes de frecuencia tanto positivas como negativas.

La transformada de Fourier, al aplicarse a la se√±al, descompone la misma en estas componentes espectrales, permiti√©ndonos ver con claridad las frecuencias que componen la se√±al y su respectiva amplitud. Esta relaci√≥n entre el dominio del tiempo y el dominio de la frecuencia es crucial para el an√°lisis de se√±ales en m√∫ltiples campos, como la ingenier√≠a el√©ctrica, la f√≠sica, y la procesamiento de se√±ales.
